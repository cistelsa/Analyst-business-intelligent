```python
import streamlit as st
from streamlit_echarts import st_echarts
import pandas as pd
import altair as alt
import numpy as np
import math
import json
from streamlit_echarts import JsCode
```

```python
df_superstore = pd.read_csv("./dataset/superstore.csv",sep=";")
df_superstore.head(10)
```

```python
def clean_data(df_superstore):
        df_superstore['Sales'] = df_superstore['Sales'].str.replace(",", ".", regex=False).astype("float")
        df_superstore['Profit'] = df_superstore['Profit'].str.replace(",", ".", regex=False).astype("float")
        return df_superstore

df_superstore_clean = clean_data(df_superstore.copy())

df_superstore_clean.head()
```

```python
def bar_data(df_bar):
    df_bar = df_bar.groupby(['Category']).agg(Ventas=('Sales', 'sum'), Ganancia=('Profit', 'sum'))
    df_bar = df_bar.sort_values(['Ventas'], ascending=[False]).reset_index()
    return df_bar
df_options = bar_data(df_superstore_clean.copy())
df_options.head()
```

```python
def click_data(df_click):
    # Change column type to datetime64[ns] for column: 'Order Date'
    df_click['Anio'] = pd.to_datetime(df_click['Order Date'], format='%d/%m/%Y')
    df_click['Anio'] = df_click['Anio'].dt.strftime('%Y-%m')  # Formato de solo año y mes (YYYY-MM)
    # Performed 2 aggregations grouped on columns: 'Anio', 'Category'
    df_click = df_click.groupby(['Anio', 'Category']).agg(Ventas=('Sales', 'sum'), Ganancia=('Profit', 'sum'))
    df_click = df_click.pivot_table(index='Anio', columns='Category', values=['Ventas', 'Ganancia'])
    return df_click

df_clicked_label = click_data(df_superstore_clean.copy())
df_clicked_label.head()
```

```python
def hot_map(df_hot_map):
    df_hot_map['Order Date'] = pd.to_datetime(df_hot_map['Order Date'], format='%d/%m/%Y')
    df_hot_map['Anio'] = df_hot_map['Order Date'].dt.strftime('%Y')
    # Performed 1 aggregation grouped on columns: 'Anio', 'Customer Name'
    df_hot_map = df_hot_map.groupby(['Anio', 'Customer Name']).agg(Sales_sum=('Sales', 'sum')).reset_index()
    return df_hot_map

df_heatmap_cartesian = hot_map(df_superstore_clean.copy())
df_heatmap_cartesian.head()
```

```python
df_grouped = df_heatmap_cartesian.groupby('Customer Name')['Sales_sum'].sum()
df_sorted = df_grouped.sort_values(ascending=False)
df_sorted.head(10)
```

```python
top_5_customers = df_sorted.head(5)
```

```python
top_5_customers_names = top_5_customers.index.tolist()
top_5_customers_names
```

```python
df_filtered_customer = df_heatmap_cartesian[df_heatmap_cartesian['Customer Name'].isin(top_5_customers_names)]
df_filtered_customer
```

```python
anios_list = sorted(set(df_filtered_customer["Anio"]))
anios_list.append('Total')
anios_list
```

```python
pivot_table = df_filtered_customer.pivot_table(index='Customer Name', columns='Anio', values='Sales_sum', aggfunc='sum', fill_value=0)
pivot_table
```

```python
pivot_table['Total'] = pivot_table['2014'] + pivot_table['2015'] + pivot_table['2016'] + pivot_table['2017'] 
pivot_table = pivot_table.sort_values(by='Total',ascending=False)
pivot_table
```

```python
desglosado_df = pd.DataFrame(pivot_table.to_records())
desglosado_df
```

```python
data = []
for customer_idx, customer in enumerate(top_5_customers_names):
    for year_idx, year in enumerate(range(2014, 2018)):
        sales = desglosado_df.loc[desglosado_df['Customer Name'] == customer, str(year)].values[0]
        data.append([customer_idx, year_idx, sales.round(2)])
    total_sales = pivot_table.loc[customer, 'Total']  # Obtener el total de ventas del cliente
    data.append([customer_idx, len(range(2014, 2018)), total_sales.round(2)])  # Agregar el total al final de cada cliente
data
```

```python
data = [[d[1], d[0], d[2] if d[2] != 0 else "-"] for d in data]
data
```

```python
def bar_Q(df_bar_q):
    df_bar_q['Fecha'] = pd.to_datetime(df_bar_q['Order Date'], format='%d/%m/%Y')
    df_bar_q = df_bar_q.groupby(['Fecha']).agg(Ventas=('Sales', 'sum')).reset_index()
    df_bar_q['Ventas'] = df_bar_q['Ventas'].round(2)
    df_bar_q['Trimestre'] = df_bar_q['Fecha'].dt.quarter
    return df_bar_q

df_q = bar_Q(df_superstore_clean.copy())
df_q.head(10)
```

```python
pivot_table = pd.pivot_table(df_q, values='Ventas', index=df_q['Fecha'].dt.year, columns='Trimestre', aggfunc='sum')
pivot_table
```

```python
matriz_q = [['Anio', 'Q1', 'Q2', 'Q3', 'Q4']]
for index, row in pivot_table.iterrows():
    fila = [str(index)]
    fila.extend(row.values.tolist())
    matriz_q.append(fila)
```

```python
def top_3(df_top_3):
    df_top_3 = df_top_3.groupby(['Product Name']).agg(conteo_ventas=('Sales', 'count'))
    df_top_3 = df_top_3.sort_values(['conteo_ventas'], ascending=[False]).reset_index()
    df_top_3 = df_top_3.set_index('Product Name')
    return df_top_3

df_bar_top_3 = top_3(df_superstore_clean.copy())
df_bar_top_3.head()
```

```python
top_3_products = df_bar_top_3.head(3)
```

```python
matriz_top_3 = [['Producto', 'No. Ventas']]
for index_top_3, row_3 in top_3_products.iterrows():
    fila_3 = [str(index_top_3)]
    fila_3.extend(row_3.values.tolist())
    matriz_top_3.append(fila_3)
```
